**Task 06:-**Calculate mean, median, and standard deviation for price and km.

```python
import pandas as pd

# Example dataset
# Load the Excel file
file_path = '/content/output.xlsx'
df = pd.read_excel(file_path)

# Convert 'km' column to numeric, removing 'km' and commas, and handling hyphens
df['km'] = df['km'].str.replace(' km', '', regex=False).str.replace(',', '', regex=False)
df['km'] = pd.to_numeric(df['km'], errors='coerce').fillna(0) # Convert to numeric, replace errors with NaN, and fill NaN with 0


# Calculate mean, median, and standard deviation for 'price' and 'km'
mean_price = df['price'].mean()
median_price = df['price'].median()
std_price = df['price'].std()

mean_km = df['km'].mean()
median_km = df['km'].median()
std_km = df['km'].std()

# Print the results
print("Price - Mean:", mean_price, "Median:", median_price, "Standard Deviation:", std_price)
print("Km - Mean:", mean_km, "Median:", median_km, "Standard Deviation:", std_km)
```

**Task 17:**-Create a new column for the age of the car using the registration field.

```python
# prompt: Create a new column for the age of the car using the registration field.

# Assuming 'df' is your DataFrame from previous code
import pandas as pd

# Load the Excel file
file_path = '/content/output.xlsx'
df = pd.read_excel(file_path)

# Convert 'registration' column to datetime objects if not already done
df['registration'] = pd.to_datetime(df['registration'], errors='coerce')

# Calculate the current year
current_year = pd.to_datetime('today').year

# Create the 'car_age' column
df['car_age'] = current_year - df['registration'].dt.year

# Handle potential errors (e.g., if registration date is invalid)
df['car_age'] = df['car_age'].fillna(0).astype(int) # Fill NaN with 0 and convert to integer

# Print the updated DataFrame
df
```

**Task 20**:-Create categories for cars based on price (e.g., low, mid, high).

```python
# prompt: Create categories for cars based on price (e.g., low, mid, high).

# Assuming 'df' is your DataFrame from the previous code
# and it contains a 'price' column.

def categorize_price(price):
    if price < 10000:
        return 'low'
    elif price < 20000:
        return 'mid'
    else:
        return 'high'

df['price_category'] = df['price'].apply(categorize_price)

print(df[['price', 'price_category']])
```

Task 26:-Group the vehicles based on attributes like price, km, and age.

```python
# prompt: Group the vehicles based on attributes like price, km, and age.

# Assuming 'df' is your DataFrame from the previous code
# and it contains 'price', 'km', and 'car_age' columns.

# Define price bins
price_bins = [0, 10000, 20000, float('inf')]
price_labels = ['Low', 'Medium', 'High']

# Define km bins
km_bins = [0, 50000, 100000, float('inf')]
km_labels = ['Low', 'Medium', 'High']

# Define age bins
age_bins = [0, 2, 5, float('inf')]
age_labels = ['New', 'Used', 'Old']

# Create new columns for categories
df['price_group'] = pd.cut(df['price'], bins=price_bins, labels=price_labels, right=False)
df['km_group'] = pd.cut(df['km'], bins=km_bins, labels=km_labels, right=False)
df['age_group'] = pd.cut(df['car_age'], bins=age_bins, labels=age_labels, right=False)

# Group vehicles based on categories
grouped_vehicles = df.groupby(['price_group', 'km_group', 'age_group'])

# Display the grouped data (you can further analyze each group)
for (price_cat, km_cat, age_cat), group_data in grouped_vehicles:
    print(f"Price: {price_cat}, KM: {km_cat}, Age: {age_cat}")
    print(group_data)
    print("-" * 20)
```
